```
#  Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
uploaded = files.upload()


#  Load dataset
df = pd.read_csv("student_marks.csv")

#  Preview data
print(df.head())

#  Check columns
print("\nColumns:", df.columns)

#  Automatically pick numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
print("\nNumeric Columns:", numeric_cols)

#  Assume:
# First numeric column = study hours
# Second numeric column = marks

hours_column = numeric_cols[0]
marks_column = numeric_cols[1]

print(f"\nUsing '{hours_column}' as Hours and '{marks_column}' as Marks")

#  Scatter plot
plt.scatter(df[hours_column], df[marks_column])
plt.xlabel(hours_column)
plt.ylabel(marks_column)
plt.title("Hours Studied vs Marks")
plt.show()

#  Correlation
correlation = df[hours_column].corr(df[marks_column])
print("\nCorrelation:", correlation)

#  Trend line
z = np.polyfit(df[hours_column], df[marks_column], 1)
p = np.poly1d(z)

plt.scatter(df[hours_column], df[marks_column])
plt.plot(df[hours_column], p(df[hours_column]))
plt.xlabel(hours_column)
plt.ylabel(marks_column)
plt.title("Trend Line")
plt.show()

avg_marks = df.groupby(hours_column)[marks_column].mean()

# Bar graph
avg_marks.plot(kind='bar')

plt.xlabel("Hours Studied")
plt.ylabel("Average Marks")
plt.title("Average Marks by Study Hours")
plt.show()
```
